Implementação em Python para transformar uma imagem colorida para níveis de cinza (0 a 255) e para binarizada (0 e 255), preto e branco

```python
import requests

image_url = "https://i.pinimg.com/originals/4e/c7/a4/4ec7a45af7ffe4f532f4fff2a2a1e855.jpg"
image_path = "/content/original_image.jpg"

response = requests.get(image_url)
if response.status_code == 200:
    with open(image_path, 'wb') as f:
        f.write(response.content)
    print(f"Imagem baixada com sucesso para: {image_path}")
else:
    print(f"Falha ao baixar a imagem. Código de status: {response.status_code}")
```

```python
from PIL import Image
import numpy as np

# Carregar a imagem original
original_image = Image.open("/content/original_image.jpg")

# Converter para níveis de cinza
grayscale_image = original_image.convert("L")

# Converter para binarizada (utilizando um threshold simples de 128)
# Você pode ajustar o threshold conforme necessário
threshold = 96
binary_image = grayscale_image.point(lambda x: 0 if x < threshold else 255, '1')

print("Imagens processadas com sucesso!")
```

```python
import matplotlib.pyplot as plt

# Exibir as imagens lado a lado
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(original_image)
axes[0].set_title("Imagem Original")
axes[0].axis('off')

axes[1].imshow(grayscale_image, cmap='gray')
axes[1].set_title("Níveis de Cinza")
axes[1].axis('off')

axes[2].imshow(binary_image, cmap='gray')
axes[2].set_title("Binarizada")
axes[2].axis('off')

plt.tight_layout()
plt.show()
```
